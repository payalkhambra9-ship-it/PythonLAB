  ### constructor

    Definition:

    A constructor is a special method in a class that is automatically called when an object of the class is created. It is used to initialize the attributes of the object.


    ### Init()

```python
class Student:
    def __init__(self,name,marks):
        self.name=name  ### instance variable
        self.marks=marks ### instance variable
st=Student(name="Rohit",marks=20) 
ss=Student(name="Rahul",marks=30)        
```

```python
class College:
    def __init__(self,name,marks):
        self.name=name  ### instance variable
        self.marks=marks ### instance variable
    def show(self):
        print("My name is",self.name)
        print("My marks are",self.marks)
st=College(name="Rohit",marks=20)            
```

```python
st.show()
```

```python
class College:
    def __init__(self,marks,college_name="liet"):
        self.college_name=college_name ### instance variable
        self.marks=marks ### instance variable
    def show(self):
        print("My college name is",self.college_name)
        print("My marks are",self.marks)
st=College(college_name="lords",marks=20)            
```

```python
st.show()
```
