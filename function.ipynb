    Function in python
    
 a function is a block of code which only runs when it is called.


    Types of function

     1. Built-in function
     
     These are the functions which are already defined in python.

     We can use these functions directly in our program.

     2. User-defined function

     These are the functions which are defined by the user.

     We can use these functions in our program.

     


```python
def add():
    a=10
    b=20
    result=a+b
    print(result)
add()    
```

```python
def greet():
    print("Hello, welcome!")
    
greet()
```

```python
def add(a, b):
    print(a + b)

add(5, 3)
```

```python
def multiply(x, y):
    return x * y

result = multiply(4, 6)
print(result)
```

```python
def check_even_odd(num):
    if num % 2 == 0:
        print("Even number")
    else:
        print("Odd number")

check_even_odd(7)
```

```python
def country(name="India"):
    print("I live in", name)

country()
country("USA")
```
