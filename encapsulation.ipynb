    #### Encapsulation in python
     
     Definition:

      Encapsulation is the bundling of data (attributes) and methods (functions) that operate on the data into a single unit, called a class. It restricts direct access to some of the object's components, which helps in maintaining the integrity of the data.

      Data hiding

      Data Security

      Controlled Access

          In python , Encapsulation is achieved using access modifiers.

          *Public members

          *Protected members

          *Private members



```python
class LietCollege:
    _college_library="College library in campus" ## protected variable
    __college_accounts=54678329 ## private
    college_name="LIET ALWAR" ### public
lc=LietCollege()    
```

```python
lc._college_library
```

```python
lc._college_library
```

```python
lc._college_accounts
```

```python
class LietCollege:
    _account=123456789
    def accountant(self):
        print(self._account)
lc=LietCollege()
lc.accountant()
        
```
