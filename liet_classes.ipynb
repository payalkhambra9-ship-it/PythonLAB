```python
### use of for  loop
class Student1:
    student_name = "gunjan"

    def doc(self):
        for i in range(5):
            print("My name is", self.student_name)

st = Student1()
st.doc()
```

```python
### use of while loop
class Student1:
    student_name = "gunjan"

    def doc(self):
        i = 1
        while i <= 3:
            print("My name is", self.student_name)
            i += 1

st = Student1()
st.doc()
```

```python
###loop through a list
class Student1:
    def doc(self):
        names = ["gunjan", "payal", "riya"]
        for name in names:
            print("Student name is", name)

st = Student1()
st.doc()
```

```python
### number loop
class Student1:
    def doc(self):
        for i in range(1, 6):
            print("Number is", i)

st = Student1()
st.doc()
```

```python
### nested loop
class Student1:
    def doc(self):
        for i in range(1, 4):
            for j in range(1, 3):
                print("i =", i, "j =", j)

st = Student1()
st.doc()
```

```python
class BankAccount:
    # Constructor
    def __init__(self, name, account_number, balance=0):
        self.name = name
        self.account_number = account_number
        self.balance = balance

    # Check balance
    def check_balance(self):
        print("Account Holder:", self.name)
        print("Account Number:", self.account_number)
        print("Current Balance: ₹", self.balance)

    # Deposit money
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print("₹", amount, "deposited successfully.")
        else:
            print("Invalid deposit amount.")

    # Withdraw money
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance!")
        elif amount <= 0:
            print("Invalid withdrawal amount.")
        else:
            self.balance -= amount
            print("₹", amount, "withdrawn successfully.")



# Create object using constructor
account = BankAccount("Priya Jain", "123456789", 1000)

# Check balance
account.check_balance()

# Deposit money
account.deposit(500)

# Check balance again
account.check_balance()

# Withdraw money
account.withdraw(300)

# Final balance check
account.check_balance()
```
