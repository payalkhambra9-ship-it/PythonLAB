#### inheritance

Inheritance is a mechanism in object-oriented programming that allows a class to inherit properties and methods from another class. The class that inherits is called the derived class or subclass, and the class being inherited from is called the base class or superclass.

         It helps in:

 Code reusability

 Reducing redundancy

 Creating a hierarchial relationship between classes

### Hybrid Inheritance

```python
# Base Class
class Grandfather:
    def show_grandfather(self):
        print("I am Grandfather")

# Multilevel Inheritance
class Father(Grandfather):
    def show_father(self):
        print("I am Father")

# Another Base Class
class Mother:
    def show_mother(self):
        print("I am Mother")

# Hybrid Inheritance (Multilevel + Multiple)
class Child(Father, Mother):
    def show_child(self):
        print("I am Child")

# Creating object
obj = Child()

obj.show_grandfather()  # From Grandfather
obj.show_father()       # From Father
obj.show_mother()       # From Mother
obj.show_child()        # From Child
```
